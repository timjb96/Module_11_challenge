```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://redplanetscience.com/"
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
#find all articles
article_divs = soup.find_all('div', class_="list_text")
print(article_divs)
```

```python
#list to containt article info- title and preview text
article_info= []
```

```python
for article in article_divs:
    title = article.find("div", class_="content_title").text
    print(title)
    summary = article.find("div", class_="article_teaser_body").text
    print(summary)
    article_dict = {}
    article_dict["Title"] = title
    article_dict["Summary"] = summary
    article_info.append(article_dict)
    
```
